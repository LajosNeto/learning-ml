# Central Limit Theorem

Central limit theorem overview :

<br>

1. If you draw large, random samples from any population, the means of those samples will be distirbuted normally around the population mean (regardeless of what the distribution pf the underlying population looks like).


2. Most sample means will lie resonably close to the population mean; the standard error is what defines "reasonably close."


3. The central limit theorem tell us the probability that a sample mean will lie within a certain distance of the population mean. It is relatively unlikely that a sample mean will lie more than two standard erros from the population mean, and extremely unlikely that it wiill lie three or more standard errors from the population mean.


4. The less likely it is that an outcome has been observed by chance, the more confident we can be in sumising that some other factor is in play.

```python
random.seed(42)

random_mean_20 = []

fig, axes = plt.subplots(5, 2, figsize=(18,30), sharex=True, sharey=True)
for i in range(10):
    rnd = random.sample(list(population_age), 20)
    mean = np.mean(rnd)
    random_mean_20.append(mean)
    sns.distplot(population_age, ax=axes[i//2,i%2])
    ax = sns.distplot(rnd, ax=axes[i//2,i%2], label='small')
    ax.legend(labels=['Random population mean : {}'.format(mean), 
                      'Original whole population mean : {}'.format(pop_mean)])
plt.show()
```

It is possible to note that for small samples, the mean deviates a lot from the whole population.

We can say that if we take a small sample from a population, this sample distirbution may not represent the true distributions from that population, with a noticiable deviation.

With a larger random sample it is possible to notice that the deviations is smaller and we have a distribution closer to the original one with the whole population.
